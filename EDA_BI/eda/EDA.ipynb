       SUM(TotalPrice) AS TotalRevenue
FROM online_retail

ORDER BY TotalRevenue DESC

SELECT InvoiceYear,

<h2>5. EXECUTE SQL FROM PYTHON</h2>

```python
def run_query(query):
    return pd.read_sql(query, conn)
```

```python
run_query("""
SELECT Description, SUM(TotalPrice) AS Revenue
FROM online_retail
GROUP BY Description
ORDER BY Revenue DESC
LIMIT 5
""")
```

```python
run_query("""
SELECT Description,
       SUM(TotalPrice) AS TotalRevenue
FROM online_retail
GROUP BY Description
ORDER BY TotalRevenue DESC
LIMIT 5
""")
```

```python
run_query("""
SELECT InvoiceYear,
       InvoiceMonth,
       SUM(TotalPrice) AS MonthlyRevenue
FROM online_retail
GROUP BY InvoiceYear, InvoiceMonth
ORDER BY InvoiceYear, InvoiceMonth
""")
```

```python
run_query("""
SELECT Description,
       COUNT(*) AS PurchaseCount
FROM online_retail
GROUP BY Description
ORDER BY PurchaseCount DESC
LIMIT 5
""")
```

```python
run_query("""
SELECT Country,
       ROUND(SUM(TotalPrice) * 100.0 / 
       (SELECT SUM(TotalPrice) FROM online_retail), 2) AS RevenuePercentage
FROM online_retail
GROUP BY Country
ORDER BY RevenuePercentage DESC
""")
```

<h1>Multivariate Analysis & Correlation</h1>

<h2>0. IMPORT LIBRARIES AND LOAD DATA</h2>

```python
# import pandas as pd
# import matplotlib.pyplot as plt
import seaborn as sns

# Load final dataset
#df = pd.read_csv("../data/final_online_retail.csv")

# Convert InvoiceDate to datetime (safety check)
df["InvoiceDate"] = pd.to_datetime(df["InvoiceDate"])

print("Dataset ready for multivariate analysis!")
```

<h2>1. CORRELATION BETWEEN NUMERICAL VARIABLES</h2>

```python
# Select numerical columns
numerical_cols = ["Quantity", "Price", "TotalPrice"]

# Correlation matrix
corr = df[numerical_cols].corr()
print(corr)

# Heatmap
plt.figure(figsize=(6,5))
sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```

<h2>2. QUANTITY VS TOTAL PRICE</h2>

```python
plt.figure(figsize=(8,6))
sns.scatterplot(data=df.sample(5000), x="Quantity", y="TotalPrice", alpha=0.5)
plt.title("Quantity vs TotalPrice")
plt.xlabel("Quantity")
plt.ylabel("TotalPrice")
plt.show()
```

<h2>3. UNIT PRICE VS TOTAL PRICE</h2>

```python
plt.figure(figsize=(8,6))
sns.scatterplot(data=df.sample(5000), x="Price", y="TotalPrice", alpha=0.5, color='orange')
plt.title("UnitPrice vs TotalPrice")
plt.xlabel("UnitPrice")
plt.ylabel("TotalPrice")
plt.show()
```

<h2>4. TOP COUNTRIES VS TOTAL REVENUE</h2>

```python
country_revenue = df.groupby("Country")["TotalPrice"].sum().sort_values(ascending=False).head(10)

plt.figure(figsize=(10,5))
sns.barplot(x=country_revenue.index, y=country_revenue.values)
plt.title("Top 10 Countries by Revenue")
plt.xlabel("Country")
plt.ylabel("Revenue")
plt.show()
```

<h2>5. MONTHLY TREND OF TOTAL REVENUE</h2>

```python
monthly_revenue = df.groupby(["InvoiceYear", "InvoiceMonth"])["TotalPrice"].sum().reset_index()
monthly_revenue["MonthYear"] = monthly_revenue["InvoiceYear"].astype(str) + "-" + monthly_revenue["InvoiceMonth"].astype(str)

plt.figure(figsize=(12,5))
sns.lineplot(x="MonthYear", y="TotalPrice", data=monthly_revenue, marker="o")
plt.xticks(rotation=45)
plt.title("Monthly Revenue Trend")
plt.xlabel("Month-Year")
plt.ylabel("Total Revenue")
plt.show()
```
